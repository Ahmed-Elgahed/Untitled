```python
import re
import os
import requests
from bs4 import BeautifulSoup

# وظيفة لاستبدال الأحرف غير الصالحة في اسم الملف
def sanitize_filename(filename):
    return re.sub(r'[<>:"/\\|?*]', "_", filename)

# تهيئة المتصفح وفتح الرابط
def scrape_page(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()  # التحقق من نجاح الطلب
        page_soup = BeautifulSoup(response.text, "html.parser")

        # استخراج اسم الملف بشكل آمن باستخدام sanitize_filename
        file_name = url.split("/")[-2] if url.split("/")[-1] == "" else url.split("/")[-1]
        file_name = sanitize_filename(file_name)  # تنظيف اسم الملف
        file_path = f"scraped_data/{file_name}.txt"
        
        # إنشاء مجلد لحفظ البيانات إذا لم يكن موجودًا
        if not os.path.exists("scraped_data"):
            os.makedirs("scraped_data")

        # حفظ النصوص في الملف
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(page_soup.get_text())

        # حفظ الصور
        images = page_soup.find_all("img")
        for img in images:
            img_url = img.get("src") or img.get("data-src")
            if img_url:
                if img_url.startswith("//"):
                    img_url = "https:" + img_url  # إصلاح الروابط الناقصة
                elif img_url.startswith("/"):
                    img_url = base_url + img_url

                img_data = requests.get(img_url, headers=headers).content
                img_name = img_url.split("/")[-1]
                with open(f"scraped_data/{img_name}", "wb") as img_file:
                    img_file.write(img_data)

        print(f"✅ تم حفظ: {url}")

    except requests.exceptions.RequestException as e:
        print(f"❌ خطأ في تحميل {url}: {e}")

# مثال لاستخدام الكود
main_url = "https://www.geeksforgeeks.org/machine-learning/?ref=header_search"
scrape_page(main_url)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import requests
import os
import time

# تهيئة المتصفح
chrome_options = Options()
chrome_options.add_argument("--headless")  # تشغيل المتصفح بدون واجهة
chrome_options.add_argument("--disable-blink-features=AutomationControlled")  # منع كشف البوت
service = Service("C:/Users/ASUS/Downloads/chromedriver-win64/chromedriver.exe")
driver = webdriver.Chrome(service=service, options=chrome_options)

# أدخل رابط الصفحة الأساسية هنا
main_url = "https://www.geeksforgeeks.org/machine-learning/?ref=header_search"

# تحميل الصفحة
driver.get(main_url)
time.sleep(3)  # تأخير بسيط لضمان تحميل المحتوى
soup = BeautifulSoup(driver.page_source, "html.parser")

# استخراج كل الروابط الداخلية
links = set()  # استخدام Set لمنع التكرار
base_url = "https://www.geeksforgeeks.org"

for a in soup.find_all("a", href=True):
    link = a["href"]
    if link.startswith("/"):  # تحويل الروابط النسبية إلى روابط كاملة
        link = base_url + link
    if base_url in link:  # تجنب الروابط الخارجية
        links.add(link)

# إنشاء مجلد لحفظ البيانات
if not os.path.exists("scraped_data"):
    os.makedirs("scraped_data")

# وظيفة لجلب وحفظ النصوص والصور من كل رابط
def scrape_page(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()  # التحقق من نجاح الطلب
        page_soup = BeautifulSoup(response.text, "html.parser")

        # حفظ النصوص مع تسمية الملفات برقم عشوائي لتجنب التكرار
        file_name = url.split("/")[-2] if url.split("/")[-1] == "" else url.split("/")[-1]
        file_path = f"scraped_data/{file_name}.txt"
        
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(page_soup.get_text())

        # حفظ الصور
        images = page_soup.find_all("img")
        for img in images:
            img_url = img.get("src") or img.get("data-src")
            if img_url:
                if img_url.startswith("//"):
                    img_url = "https:" + img_url  # إصلاح الروابط الناقصة
                elif img_url.startswith("/"):
                    img_url = base_url + img_url

                img_data = requests.get(img_url, headers=headers).content
                img_name = img_url.split("/")[-1]
                with open(f"scraped_data/{img_name}", "wb") as img_file:
                    img_file.write(img_data)

        print(f"✅ تم حفظ: {url}")

    except requests.exceptions.RequestException as e:
        print(f"❌ خطأ في تحميل {url}: {e}")

# استخراج البيانات من الصفحة الرئيسية والروابط الداخلية
scrape_page(main_url)

for link in list(links)[:5]:  # تقليل عدد الصفحات التي يتم تحميلها (يمكن زيادته)
    scrape_page(link)

driver.quit()
print("🎉 تم حفظ جميع البيانات بنجاح!")
```
